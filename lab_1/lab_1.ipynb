```python
import pandas as pd
from config import PROJECT_PATH
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(8, 8)})
```

```python
def plot_0(df, x_arg):
    sns.set_theme(style="ticks", palette="pastel")
    
    g = sns.boxplot(
        x=x_arg,
        y="f_score",
        data=df
    )
    g.set_yscale("logit")
    
    sns.despine(offset=10, trim=True)
    plt.show()
```

```python
def plots_for_dataset(df, dataset_type):
    df_tmp = df.loc[df['dataset_type'] == dataset_type]

    x_args = list(df_tmp.columns)
    x_args.remove('dataset_type')
    x_args.remove('f_score')

    df_tmp = df_tmp[x_args]
    df_tmp = df.groupby(x_args, as_index=False).mean()
    df_std = df.groupby(x_args, as_index=False).std()
    df_tmp['f_score_std'] = df_std['f_score']

    for x_arg in x_args:
        plot_0(df, x_arg)
```

```python
df = pd.read_csv('{}/lab_1/results.csv'.format(PROJECT_PATH))
```

```python
plots_for_dataset(df.copy(), 'iris')
```

```python
plots_for_dataset(df.copy(), 'wine')
```

```python
plots_for_dataset(df.copy(), 'glass')
```
